```python
import networkx as nx
import matplotlib.pyplot as plt
```

From http://www-personal.umich.edu/~mejn/netdata/

```python
H = nx.read_gml("power.gml", label = 'id')
```

remove any node > 500 to shrink the network

```python
for i in range(250, 4941):
    H.remove_node(i)
```

https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.components.connected_components.html#networkx.algorithms.components.connected_components

```python
largest_cc = max(nx.connected_components(H), key=len)
largest_subgraph = H.subgraph(largest_cc).copy()
    
```
